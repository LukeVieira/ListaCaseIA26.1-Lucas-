## Lista de Exercícios: Revisão Teórica do Módulo 03

### Módulo 3.1: KNN (K-Nearest Neighbors)

**Questão 1:** O KNN é um algoritmo baseado em instâncias. Explique o que significa a classificação por "votação dos vizinhos mais próximos" e como o valor de K afeta a suavidade da fronteira de decisão.

**R:** A classificação por votação dos vizinhos mais próximos significa que dado um conjunto de dados de treinamento, o algoritmo processa uma nova amostra calculando a distância Euclidiana entre ela e todos os demais pontos do conjunto. A partir desse cálculo, ele identifica os K pontos mais próximos a ele e em seguida a nova amostra é atribuída à classe mais frequente (a moda) entre esses vizinhos. Vale ressaltar que o valor de K define a suavidade da fronteira de decisão: quanto menor o valor de K, mais complexo e sensível a ruídos o modelo se torna, por outro lado, quanto maior o K, mais suave e generalista será a fronteira.


**Questão 2:** Por que a Normalização dos dados é considerada um passo obrigatório para o KNN? O que acontece com a influência das variáveis se uma tiver escala de 0 a 1 e outra de 0 a 10.000?

**R:** A normalização é considerada obrigatória para o KNN pois esse algoritmo toma decisões baseadas na distância Euclidiana entre os pontos no espaço de características. Se os dados não estiverem normalizados  as variáveis com escalas de 0 a 10.000 terão total influência na equação de distância ao contrário das variáveis de 0 a 1 que terão uma influência tão mínima nos cálculos que chegam no ponto de serem praticamente ignoradas pelo algoritmo.

### Módulo 3.2: Naive Bayes

**Questão 3:** O Naive Bayes é baseado no Teorema de Bayes e na "hipótese de independência condicional". Explique essa hipótese e discuta por que ela é considerada "ingênua" (naive).

**R:** A hipótese da independência condicional pressupõe que todas as variáveis preditoras não possuem nenhuma relação umas com as outras desde que a classe seja conhecida, ou seja, elas são consideradas como eventos matemáticamente isolados. Essa hipótese é tida com ingênua pois essa premissa raramente se faz presente na realidade.

**Questão 4:** O notebook apresenta as variantes Gaussiana, Multinomial e Bernoulli. Qual delas você escolheria para um problema de classificação de texto (como filtro de spam) baseado na frequência das palavras? Justifique.

**R:** Para problemas envolvendo classificação de texto baseado na frequência das palavras eu escolheria variante Multinomial, pois ela foi desenvolvida para operar com features discretas que representam contagens ou frequências.

### Módulo 3.3: Regressão Logística

**Questão 5:** Explique o papel da Função Sigmoide na Regressão Logística. Como ela transforma o resultado de uma equação linear em uma probabilidade entre 0 e 1?

**R:**  A função Sigmoide atua como uma transformação matemática, ela recebe o valor contínuo do cálculo linear e o "achata", mapeando-o estritamente para o intervalo numérico entre 0 e 1. Dessa forma, o resultado passa a representar matematicamente a probabilidade daquela amostra pertencer à classe positiva.

**Questão 6:** No contexto de otimização, o que é o Gradiente Descendente e como ele é utilizado para ajustar os coeficientes da Regressão Logística durante o treinamento?

**R:** O gradiente descendente é o algoritmo de otimização encarregado de minimizar a função de custo do modelo  para encontrar os melhores coeficientes. Ele atua iterativamente, atualizando os parâmetros na direção oposta ao gradiente da função de custo.


### Módulo 3.4: SVM (Support Vector Machines)

**Questão 7:** O objetivo principal do SVM é encontrar o "hiperplano ótimo". O que define esse hiperplano e como o conceito de Margem (distância entre o hiperplano e os pontos mais próximos) influencia a capacidade de generalização do modelo?

**R:** Um hiperplano ótimo é aquele que está posicionado de forma a garantir a maior distância possível entre as classes (a margem máxima). Maximizar essa margem implica em uma alta robustez na generalização para novos dados e na redução do risco de overfitting.

**Questão 8:** Explique o que são os Vetores de Suporte e por que o modelo SVM depende apenas desses pontos específicos para definir a fronteira de decisão, ignorando os demais pontos do conjunto de treinamento.

**R:** Vetores de suporte são os pontos do conjunto de treinamento mais próximos ao hiperplano e o SVM depende apenas destes pontos pois eles  definem a inclinação e a posição do hiperplano, os outros pontos são ignorados pois não afetam a equação matemática que posiciona a margem.

**Questão 9:** O que é o Kernel Trick? Como essa técnica permite que o SVM crie fronteiras de decisão não-lineares sem a necessidade de calcular explicitamente as coordenadas dos dados em um espaço de maior dimensão?

**R:** O Kernel trick é um truque matemático que permite resolver problemas não-lineares através de transformações implícitas. Ele funciona mapeando implicitamente as amostras originais para um ambiente virtual de maior dimensão onde elas passam a ser separáveis por um hiperplano reto.

### Módulo 3.5: Árvores de Decisão

**Questão 9:** As Árvores de Decisão utilizam um processo chamado Particionamento Recursivo. Explique como o algoritmo escolhe qual feature utilizar para realizar a primeira divisão (nó raiz) e as subsequentes.

**R:** O algoritmo utiliza uma estratégia "gulosa",  ele escolhe a melhor feature (aquela que mais reduz impureza) e divide os dados baseados nessa feature e repete esse processo recursivamente para cada subconjunto.

**Questão 10:** Diferencie os critérios de impureza Gini e Entropia. Em que situação o uso do Ganho de Informação (Information Gain) é determinante para a construção da árvore?

**R:** O critério Gini mede a probabilidade de classificação incorreta ao escolher um elemento aleatoriamente e o critério Entropia mede o grau de desordem ou incerteza nos dados. Ele ele determina matematicamente a construção, a árvore avalia a impureza total antes de uma possível divisão e subtrai a impureza após a divisão.

**Questão 11:** Árvores de decisão sem controle de crescimento tendem a sofrer de Overfitting. Cite três hiperparâmetros do Scikit-Learn (ex: max_depth) que podem ser ajustados para controlar a complexidade da árvore e melhorar sua generalização.

**R:** Os 3 hiperparâmetros podem ser: max_depth, o min_samples_split e o min_samples_leaf.

**Questão 12:** Qual é a principal vantagem das Árvores de Decisão em relação a modelos "caixa-preta" como Redes Neurais ou SVM no que diz respeito à interpretabilidade das decisões de negócio?

**R:** A principal vantagem é que o modelo como um fluxograma de decisões, ou seja, cada critério tomado pelo algoritmo pode ser extraído, o que torna sua adequação e interpretabilidade indispensáveis para validações de negócios.

### Módulo 3.6: Métodos de Ensemble

**Questão 13:** O conceito de Ensemble baseia-se na "sabedoria das multidões". Explique a intuição por trás de combinar vários modelos "fracos" para criar um modelo forte e como isso ajuda a reduzir o erro total (viés + variância).

**R:** Usando o exemplo da árvore de decisão dado no material, a intuição é que: uma única árvore pode, facilmente, superajustar aos dados de treinamento e cometer erros enormes, mas quando se agrupam múltiplos modelos independentes que são "fracos", seus deslizes aleatórios individuais acabam sendo diluídos e anulados. A decisão agregada diminui expressivamente a variação  das predições, alcançando respostas mais consistentes e com bem menos ruídos isolados.

**Questão 14:** Diferencie as técnicas de Bagging e Boosting. Como a forma de treinamento dos modelos (independente vs. sequencial) define cada uma dessas abordagens?

**R:** No Bagging, os modelos preditivos são estruturados operando de maneira sozinha e em paralelo, o algoritmo gera múltiplas versões amostrais misturadas dos mesmos dados para criar os estimadores e pondera o resultado da votação. Por outro lado, o Boosting tem sua arquitetura baseada em sequência: um modelo secundário só é treinado com o objetivo  de penalizar, isolar e focar nos erros falhos reportados e processados do modelo anterior que acabou de atuar.

**Questão 15:** A Random Forest é uma evolução do Bagging que adiciona uma camada extra de aleatoriedade na escolha das features. Por que essa seleção aleatória de subconjuntos de variáveis é importante para descorrelacionar as árvores do ensemble?

**R:** A seleção aleatória serve para forçar as árvores a serem diferentes umas das outras (descorrelacionadas). Se todas as árvores pudessem olhar para as mesmas variáveis, elas tomariam as mesmas decisões e cometeriam  os mesmos erros. Ao sortear as variáveis, a Random Forest garante que cada árvore aprenda um único padrão, fazendo com que os erros individuais se anulem na hora da votação final.

**Questão 16:** O que é o Stacking (Empilhamento) e como o papel do "Meta-Modelo" difere dos modelos base na hora de realizar a previsão final?

**R:** O Stacking é um método que usa vários modelos diferentes (como KNN, Naive Bayes e SVM) para analisar os dados e fazer previsões. A diferença do "Meta-Modelo" é que ele não olha para os dados originais do problema. Ele analisa apenas as "opiniões" dadas pelos outros modelos e aprende em qual deles ele deve confiar mais, juntando essas respostas para fazer a decisão final.

### Módulo 4.1: Regressão Linear Múltipla

**Questão 1:** A Regressão Linear Múltipla possui uma solução analítica fechada baseada no Método dos Mínimos Quadrados. Apresente matematicamente a Equação Normal utilizada para encontrar o vetor de coeficientes $\hat{\beta}$ e explique brevemente o que representam as matrizes $X$ e $Y$ nessa formulação.

**R:** A Equação Normal utilizada para encontrar o vetor de coeficientes $\hat{\beta}$ é dada por $$\hat{\beta} = (X^T X)^{-1} X^T Y ,$$ e nessa formulação matricial a matriz $X$ (matriz de características) é a que contém todas as variáveis independentes e a matriz $Y$(vetor alvo) é o vetor coluna que contém a variável dependente.

**Questão 2:** Na regressão linear múltipla, cada feature recebe um coeficiente $\beta_i$. Como deve ser interpretado o valor desse coeficiente no contexto da predição? Além disso, explique o que é o fenômeno da multicolinearidade e por que ele pode distorcer a interpretação matemática desses pesos (como visto no exemplo prático entre número de quartos e renda).

**R:** O coeficiente $\beta_i$ indica o impacto direto daquela feature: se ela valer 10, a sua predição aumenta em 10 unidades para cada vez que aquela feature subir 1 ponto, assumindo que as outras características fiquem paradas. A multicolinearidade acontece quando duas ou mais variáveis estão dizendo praticamente a mesma coisa, isso confunde a matemática do algoritmo, que não consegue mais separar de quem é o mérito da predição, distorcendo os pesos e tirando o sentido lógico da interpretação isolada.

**Questão 3:** Ao implementar a Regressão Linear Múltipla do zero ou ao utilizar o Scikit-Learn, o algoritmo busca otimizar uma função de custo específica. Qual é a principal métrica de erro minimizada durante a fase de treinamento e como ela se difere, em propósito, do coeficiente de determinação ($R^2$)?

**R:**  A métrica de erro minimizada no treinamento é o Erro Quadrático Médio (MSE), que mede a "distância física" entre as previsões e a realidade. O seu propósito é ser uma penalidade a ser reduzida. O coeficiente de determinação ($R^2$) não mede distância de erro, mas sim a qualidade do modelo em porcentagem: ele indica qual a proporção (de 0 a 1) da variação do seu alvo que o modelo conseguiu compreender e explicar.

### Módulo 4.2: Support Vector Regression (SVR)

**Questão 4:** O SVR adapta os conceitos tradicionais de margem do SVM (usado para classificação) para o cenário de predição de valores contínuos. Explique o conceito do "tubo $\epsilon$" (epsilon) e como ele altera a forma como o modelo penaliza os erros de previsão em comparação com a minimização do erro quadrático tradicional da Regressão Linear.

**R:** O "tubo $\epsilon$" (epsilon) é uma zona de tolerância que o SVR desenha ao redor da sua linha de previsão. A diferença para a regressão linear é que o SVR  ignora  qualquer erro (ponto de dado) que caia dentro da largura desse tubo. Ele só se importa em penalizar matematicamente os erros grandes que vazam para fora da margem.

Questão 5: O SVR é capaz de modelar relações não lineares complexas através do *Kernel Trick*. O que é essa técnica matemática e como ela permite que o algoritmo encontre hiperplanos de regressão curvados sem a necessidade de calcular explicitamente as coordenadas dos dados em um espaço de dimensões superiores?

**R:** O Kernel Trick é um truque matemático que traduz dados complexos para um espaço virtual de mais dimensões, onde eles se tornam separáveis por um plano reto. Isso permite que o algoritmo construa modelos de previsão curvos e não lineares de maneira rápida, pois ele calcula as distâncias usando fórmulas prontas (Kernels) em vez de gastar processamento tentando descobrir a posição 3D de cada dado.

Questão 6: No treinamento de um modelo SVR, os hiperparâmetros $C$ e $\epsilon$ desempenham papéis cruciais de regularização. Discuta como o aumento ou a diminuição de cada um desses dois parâmetros afeta a flexibilidade (suavidade) do modelo e a quantidade resultante de vetores de suporte.

 **R:**  O $C$ controla a punição para os pontos fora do tubo: um $C$ alto não tolera erros (modelo mais rígido e com chance de superajuste), e um $C$ baixo aceita erros (modelo mais flexível e suave). Já o $\epsilon$ dita o tamanho do tubo: um tubo grande ($\epsilon$ alto) perdoa muito e engloba a maioria dos pontos, precisando de poucos vetores de suporte para se ancorar, um tubo estreito ($\epsilon$ baixo) exige alta precisão e transforma muitos pontos em vetores de suporte.

### Módulo 4.3: Árvores de Regressão

**Questão 7:** Diferentemente das Árvores de Classificação que utilizam métricas como Gini ou Entropia para avaliar a impureza de um nó, as Árvores de Regressão utilizam métricas de dispersão. Explique como o algoritmo de particionamento recursivo utiliza a Soma dos Erros Quadráticos (SSE) para definir qual *feature* e qual ponto de corte proporcionam a melhor divisão.

**R:** Enquanto a classificação usa Gini para separar categorias, a regressão testa os cortes tentando encontrar aquele que deixe os números dentro de cada agrupamento o mais parecidos possível. O algoritmo faz isso calculando a Soma dos Erros Quadráticos (SSE), a divisão escolhida será sempre aquela que entregar o menor SSE matemático no final do corte.

**Questão 8:** A predição de uma Árvore de Regressão possui um formato visual em "degraus". Quando uma nova observação percorre os nós da árvore e atinge uma folha final específica, como o valor numérico exato da predição é matematicamente calculado para essa amostra (considerando os dados de treino que caíram ali)?

**R:** Ele é matematicamente calculado da seguinte forma: quando um dado novo chega, a predição exata dele é o cálculo da média aritmética de todos os números reais de treinamento que pararam dentro daquela mesma folha.

**Questão 9:** Árvores de regressão que crescem sem restrições são altamente propensas ao overfitting, memorizando o ruído dos dados. Cite e explique o funcionamento de pelo menos dois hiperparâmetros (como `max_depth` ou `min_samples_leaf`) utilizados para controlar o crescimento da árvore e garantir sua capacidade de generalização.

**R:** O max_depth coloca um teto para o tamanho da árvore, proibindo que ela continue se dividindo infinitamente. O min_samples_leaf exige que a folha final tenha uma população mínima. Juntos, eles impedem que o modelo crie caminhos absurdos e folhas solitárias apenas para memorizar uma única casa fora da curva (ruído).

### Módulo 4.4: Métodos Ensemble para Regressão

**Questão 10:** O Random Forest Regressor é considerado uma evolução direta do método de Bagging tradicional. Qual é a restrição aleatória específica introduzida pelo Random Forest durante o processo de divisão dos nós das árvores, e por que essa técnica é fundamental para descorrelacionar os modelos base e reduzir a variância do *ensemble*?

**R:** A restrição da Random Forest consiste em esconder algumas variáveis e forçar o algoritmo a escolher o melhor corte baseando-se apenas num sorteio aleatório de colunas. Isso é fundamental porque descorrelaciona as árvores, impedindo que todas fiquem fixadas pela mesma variável principal e acabem tomando as mesmas decisões.

**Questão 11:** Ao contrário de algoritmos baseados em Bagging (que treinam modelos de forma independente e paralela), o Gradient Boosting opera de forma puramente sequencial. Explique a intuição fundamental por trás do treinamento do Gradient Boosting e qual é a variável alvo matemática (o target) que cada nova árvore do conjunto tenta prever e corrigir.

**R:** O Gradient Boosting opera com foco em corrigir falhas. A primeira árvore tenta prever o preço real, a partir daí, as próximas árvores nunca olham para o preço real. A variável alvo dela passa a ser o resíduo (o erro numérico) que a árvore anterior deixou para trás, fazendo pequenas correções somadas até chegar na resposta final.

**Questão 12:** Para combinar o poder preditivo de algoritmos de naturezas completamente distintas (como Árvores, SVR e Regressão Linear), podemos utilizar métodos de Voting ou Stacking. Qual é a diferença conceitual e arquitetural na forma como o *Voting Regressor* e o *Stacking Regressor* consolidam as predições individuais dos modelos base para gerar o resultado final?

**R:** O Voting Regressor recolhe as predições de todos os modelos e tira uma média aritmética simples para fazer a decisão final. O Stacking Regressor é hierárquico: ele pega as predições dos modelos e entrega para um "Meta-Modelo"  que aprende matematicamente em quem ele deve confiar mais e qual o peso deve dar para a opinião de cada algoritmo antes de dar o resultado final.
