## Módulo 1.1: Fundamentos de IA e Machine Learning

**Questão 1:**  O termo "Inteligência Artificial" abrange muito mais do que os populares modelos de linguagem grandes. Com base no material de estudo, defina formalmente o que é a Inteligência Artificial e explique por que ela é considerada um campo amplo.

**R:** Com base no material de estudos, IA é a capacidade de máquinas realizarem tarefas que historicamente exigiam inteligência humana, ou seja, é qualquer técnica que permita a computadores simularem inteligência. Essa é uma área bem ampla pois existem várias formas e técnicas que atigem esse objetivo, desde sistemas simples baseados em regras lógicas até redes neurais complexas que aprendem padrões sofisticados em dados.

**Questão 2:** A ciência dividiu a Inteligência Artificial em diversas subáreas para otimizar o seu estudo. Liste as quatro principais subáreas mencionadas no texto e descreva brevemente o foco de atuação de cada uma delas.

**R:** As 4 principais subárea da IA são: Machine Learning onde o computador aprende padrões à partir dos dados, Deep Learning que utiliza redes neurais artificiais profundas inspiradas no cérebro humano que têm muitas camadas, Processamento de Linguagem Natural onde máquinas entendem e geram texto e Visão Computacional onde as máquinas "enxergam" e interpretam imagens.

**Questão 3:** É comum haver confusão entre os termos Inteligência Artificial (IA), Machine Learning (ML) e Deep Learning (DL). Discorra sobre a relação hierárquica entre esses três conceitos, explicando como eles se diferenciam em sua abordagem para a resolução de problemas.

**R:** Pensando na abordagem de conjuntos, IA seria o maior conjunto que engloba todas as técnicas que permitem computadores simularem inteligência, enquanto Machine Learning e Deep Learning são os subconjuntos, técnicas específicas de IA que são utilizadas em contextos diferentes. Equanto ML inclui técnicas estatísticas que permitem as máquinas melhorarem tarefas através da experiência o DL é o subconjunto do ML, onde permite que o software treine a si mesmo para desempenhar tarefas, só que se utilizando de múltiplas camadas de redes neurais artificiais.

## Módulo 1.2: Tipos de Aprendizado de Máquina

**Questão 4:** O Aprendizado Supervisionado é uma das estratégias mais comuns em Machine Learning. Descreva o funcionamento desse paradigma e explique qual é a função essencial dos "rótulos" (labels) durante a fase de treinamento do modelo.

**R:** O método de aprendizado supervisionado consiste no treinamento de algoritmos através do fornecimento de "rótulos corretos", onde o modelo aprende a relação de entrada e sáida através dos exemplos. A cada tentativa é feito um feedback sobre os erros e o modelo se ajusta. Após vários ciclos, o modelo consegue prever bem mesmo em dados novos. A função essencial dos rótulos é servir de exemplo, é através deles que o algoritmo identifica padrões.

**Questão 5:** Em oposição aos métodos supervisionados, o Aprendizado Não Supervisionado atua sobre dados brutos. Explique como esse tipo de algoritmo opera na ausência de respostas prévias e aponte qual é o principal desafio analítico ao se interpretar os resultados gerados.

**R:** Sabendo que esse tipo de algoritmo lida apenas com dados brutos, não tendo qualquer classificação prévia, ele tenta descobrir por si só padrões e estruturas. O principal problema é conseguir diferenciar se o que o agoritmo encontrou é realmente um padrão ou apenas ruído.

**Questão 6:** O Aprendizado por Reforço utiliza uma abordagem baseada na interação. Detalhe o ciclo de aprendizado de um agente neste cenário, destacando o papel das recompensas e punições na formulação de sua estratégia (política).

**R:** O ciclo de aprendizado tem as seguintes etapas:

* O agente (sistema de IA) executa ações em um ambiente
* O ambiente responde a essas ações com uma punição ou uma recompensa
* O agente assim, vai aprendendo qual sequência de ações maximiza as recompensas
* Desenvolve estratégias para ganhar mais recompensas

 As recompensas mostram quais ações são favorecidas ou não, fazendo com que o agente tome decisões e formule estratégias estritamente visando os feedbacks positivos (recompensas) e evitando qualquer feedback negativo (punições)

## Módulo 1.3: Matemática para Machine Learning

**Questão 7:** Antes de aplicar qualquer algoritmo, é necessário compreender a natureza das informações analisadas. Defina matematicamente o conceito de "dados" e como eles representam um conjunto de observações, denotadas por $X = \{x_1,x_2,...,x_n\}$.

**R:** Matematicamente, os dados são definidos como um conjunto de observações representado por uma n-upla ordenada, ou seja, $X = \{x_1, x_2, \dots, x_n\}$. Nessa notação, cada elemento $x_i$ corresponde a uma observação individual dentro desse conjunto.

**Questão 8:** As características são os blocos de construção de qualquer conjunto de dados. Apresente a definição formal de "características" e demonstre, por meio de um exemplo genérico, como uma observação $x_i$ é composta por múltiplos atributos $x_{ij}$.

**R:** As características são definidas formalmente como os atributos que utilizamos para descrever cada observação. Para exemplificar esse conceito, considere a observação $x_i = (x_{i1}, x_{i2}, \dots, x_{id})$: nela, cada valor $x_{ij}$ representa um atributo específico da observação $x_i$ (como idade, peso ou altura, por exemplo), sendo que o índice $j$ varia de $1$ até $d$.

**Questão 9:** A visualização geométrica dos dados é uma ferramenta teórica poderosa. Conceitue o "espaço de características" e explique o que significa matematicamente representar esse ambiente através da notação $R^d$.

**R:** O "espaço de características" é o conjunto de todas as possíveis combinações dos atributos utilizados para descrever uma observação. A representação matemática significa que Se cada observação é descrita por $d$ características, esse espaço contém todas as combinações possíveis desses atributos e é matematicamente representado por $R^d$, isso significa dizer que $d$ é o número de dimensões do conjunto que contém essas combinações.

**Questão 10:** Para localizar uma observação dentro do espaço de características, utilizamos estruturas matemáticas específicas. Defina o que é um vetor de características e discorra sobre como a representação $x_i = (x_{i1},x_{i2},...,x_{id})$ traduz a posição exata de um dado neste espaço.

**R:** Um vetor de características é um ente matemático utilizada para representar cada observação dentro do espaço de características. A representação $x_i = (x_{i1}, x_{i2}, \dots, x_{id}) \in R^d$ mostra a posição de um dado ao estabelecer que $i$ indica o número da observação e cada índice $j$ representa o valor de uma característica correspondente. Geométricamente esse ente pode ser visualizado como uma "seta" que parte da origem do espaço de características e aponta diretamente para as coordenadas exatas da pessoa (ou observação).

## Módulo 2.1: Modelos e Validação

**Questão 1:** (Categorias de modelos preditivos): Os modelos preditivos podem ser divididos em diferentes categorias baseadas em sua abordagem de aprendizado. Diferencie os modelos estatísticos/probabilísticos dos modelos baseados em particionamento de dados, citando a principal característica e um exemplo de cada.

**R:** 

* Modelos Estatísticos e Probabilísticos: Sua principal característica é assumir que os dados seguem uma distribuição probabilística específica, ajustando seus parâmetros com o objetivo de maximizar a verossimilhança dos dados observados. Um exemplo dessa categoria é a Regressão Linear.

* Modelos Baseados em Particionamento de Dados: A principal característica desse modelo é criar regras de decisão para dividir os dados em subconjuntos mais homogêneos, organizando-os de forma hierárquica. Um exemplo dessa categoria são as Árvores de Decisão.

**Questão 2:** (Trade-off Bias-Variância): O desempenho de um modelo preditivo é fortemente influenciado pelo equilíbrio entre viés (bias) e variância. Explique o que caracteriza um modelo com alta variância e como esse comportamento se relaciona com o fenômeno de overfitting (superajuste).

**R:** Um modelo com alta variância se caracteriza por sua grande sensibilidade às flutuações dos dados de treinamento. São modelos mais complexos que acabam aprendendo não apenas os padrões fundamentais, mas também o ruído específico dos dados de treino. Isso resulta em um excelente desempenho nos dados de treino, mas um desempenho ruim em novos dados, caracterizando o superajuste (overfitting).

**Questão 3:** (Validação cruzada): A estratégia de validação deve respeitar a natureza dos dados. Explique por que a validação cruzada K-Fold aleatória tradicional não é adequada para séries temporais e descreva como a abordagem "rolling window" (janela deslizante) resolve esse problema.

**R:** A validação K-Fold aleatória não é adequada para séries temporais porque pode causar vazamento temporal em situações onde existe forte dependência temporal (como autocorrelação, tendência e sazonalidade). Como o K-Fold tradicional muitas vezes embaralha os dados antes de criar os blocos (folds), ele ignora a linha do tempo.A abordagem rolling window (janela deslizante) resolve isso utilizando partições que respeitam a ordem cronológica, garantindo que o período de treino seja sempre anterior ao período de teste. Ela faz isso mantendo uma janela fixa de comprimento $W$, de modo que o modelo treina no intervalo $[t-W+1, t]$ e valida no período seguinte $(t, t+h]$.

## Módulo 2.2: Métricas de Classificação

**Questão 4 (Tipos de classificação):** Em problemas de classificação, as categorias (labels) podem assumir diferentes estruturas. Qual é a diferença fundamental entre a classificação multiclasse e a classificação multirótulo (multi-label) no que diz respeito à exclusividade das classes?

**R:** A diferença fundamental está na exclusividade das categorias. Na classificação multiclasse, as class labels são mutuamente exclusivas, o que significa que uma observação só pode pertencer a uma única categoria. Por outro lado, na classificação multirótulo, as labels não são mutuamente exclusivas, o que permite que uma mesma observação seja mapeada para duas ou mais categorias simultaneamente.

**Questão 5 (O que é um classificador):** Para realizar previsões de categorias, utilizamos classificadores. Defina matematicamente/algoritmicamente o que é um classificador no contexto de Machine Learning e qual é a sua função principal sobre um vetor de características.

**R:** Algoritmicamente, um classificador é um conjunto de regras (que implementa métodos matemáticos e estatísticos) que define a metodologia utilizada pelo modelo para produzir o resultado esperado. Sua função principal é atuar sobre os dados recebidos para mapear cada vetor de características a uma categoria específica.

## Módulo 2.3: Métricas de Regressão

**Questão 7 (Problemas de regressão):** Em contraste com os problemas de classificação, a regressão possui um objetivo diferente para a variável alvo. Defina qual é o foco de um problema de regressão e forneça um exemplo prático (como o citado no material sobre previsão de salário).

**R:** O foco de um problema de regressão é fazer a previsão de uma variável contínua (ou do tipo real/float), diferentemente da classificação, que prevê variáveis discretas ou categóricas. Um exemplo prático citado no material é a previsão do salário de uma pessoa utilizando como base um vetor de características que inclui sua idade, anos de educação, área de trabalho, entre outros.

**Questão 8 (Métricas específicas):** Diversas métricas são utilizadas para avaliar a precisão de um regressor. Descreva a diferença conceitual e matemática entre o Erro Médio Absoluto (MAE) e o Erro Quadrático Médio (MSE), explicando como cada um lida com a magnitude dos erros.

**R:** A diferença conceitual e matemática entre as duas métricas se dá da seguinte forma:

Erro Médio Absoluto (MAE): Conceitualmente, ele mede a diferença média entre o valor predito pelo modelo e o valor real observado, calculando o módulo dessa diferença para lidar com valores positivos e negativos. Matematicamente, é definido por:$$MAE(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i| $$ Em relação à magnitude, o MAE atribui o mesmo peso para cada erro, o que faz com que ele mantenha a escala original dos dados e não seja muito afetado por outliers.

Erro Quadrático Médio (MSE): Conceitualmente, ele também calcula o erro médio, mas eleva a diferença entre o valor predito e o real ao quadrado. Sua fórmula matemática é:$$MSE(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$Como ele eleva a diferença ao quadrado, a forma como lida com a magnitude dos erros é penalizando valores que sejam muito diferentes do real (erros maiores), o que acaba distorcendo a escala original da métrica

**Questão 9 (Interpretação de erros):** O Erro Percentual Absoluto Médio (MAPE) e o coeficiente de determinação $(R^2)$ fornecem visões complementares sobre o modelo. Como você interpretaria os resultados de um modelo preditivo de preços de imóveis que apresenta um MAPE de 12\% e um $R^2$ de 85%?

**R:** A interpretação para um modelo de precificação de imóveis seria:

* MAPE de 12%: Como o MAPE indica a porcentagem de erro em relação aos valores reais, um resultado de 12% significa que o modelo faz, em média, previsões que diferem em 12% do valor real dos imóveis.

* $R^2$ de 85%: O $R^2$ representa o percentual de variância (poder explicativo) capturado pelo modelo. Um valor de 85% significa que o seu modelo consegue explicar 85% da variação dos preços dos imóveis.

## Módulo 2.4: Feature Engineering

**Questão 10 (Criação e transformação de features):** A engenharia de features muitas vezes requer a criação de novas variáveis para revelar padrões ocultos. Cite pelo menos três exemplos de métodos ou combinações que podem ser aplicados para criar novas features a partir de dados temporais ou numéricos.

**R:** Alguns exemplos de métodos ou combinações que podem ser aplicados são:

 * Transformações: Aplicação de operações matemáticas diretas na variável, como logaritmo, extração de raiz ou padronização.

 * Interações simples: Criação de novas variáveis através da combinação de outras, como calcular o produto ou a razão entre duas variáveis.

 * Extração e agregações temporais: Para dados temporais, é possível extrair componentes específicos da data (como mês ou dia da semana) ou calcular agregações como médias móveis.

**Questão 11 (Transformação de categóricas / Encoding):** Modelos estatísticos exigem que variáveis de texto sejam convertidas em linguagem matemática. Como funciona o método de One-hot encoding e qual problema (relacionado à redundância) devemos evitar ao utilizá-lo?

**R:** O método de One-hot encoding funciona criando uma nova coluna binária (também chamada de variável dummy) para cada categoria existente dentro da variável original. O algoritmo atribui o valor numérico 1 se a observação pertencer àquela categoria específica e 0 caso contrário. O problema que devemos evitar ao utilizá-lo é a multicolinearidade. Esse problema ocorre quando uma das colunas criadas pode ser perfeitamente deduzida ou prevista a partir da soma das outras (redundância), o que pode tornar os cálculos instáveis para certos modelos estatísticos, como a regressão linear.

**Questão 12 (Tratamento de dados faltantes):** Lidar com valores nulos é uma etapa crucial no processamento de dados. Explique o que é a imputação de valores ausentes, dê dois exemplos de estratégias de preenchimento e justifique por que esse procedimento deve ser feito apenas no conjunto de treino.

**R:** A imputação de valores ausentes é o procedimento de preencher lacunas na base de dados quando certas informações não foram registradas. Dois exemplos de estratégias de preenchimento são:

 * Utilizar a média ou mediana para preencher dados ausentes em variáveis numéricas.

 * Utilizar a moda para preencher os valores ausentes em variáveis categóricas.

Esse procedimento deve ser realizado apenas no conjunto de treino (ou dentro de cada fold) para evitar o vazamento de informação aos dados de teste, garantindo que seja possível validar o impacto das métricas no modelo de Machine Learning.

**Questão 13 (Normalização e padronização):** Ajustar a escala dos dados evita que os algoritmos sejam enviesados por grandezas diferentes. Qual é a diferença prática entre aplicar um Scaling (como o Min-Max) em oposição à Padronização (Standardization), e qual delas altera a distribuição dos dados para ter média 0 e desvio padrão 1?

**R:** A diferença prática está na forma como os dados são ajustados e em como lidam com os valores extremos. O Scaling (como o Min-Max) é focado apenas em transformar os valores numéricos para que fiquem contidos dentro de um mesmo intervalo de escala, mas pode ser sensível aos valores extremos da base de dados. Em contraste, a Padronização (Standardization) foca em alterar a distribuição da variável, sendo uma opção melhor quando há presença de outliers, pois ela é menos afetada por esses valores extremos em comparação ao Min-Max. O método que altera especificamente a distribuição dos dados de modo que ela passe a ter média 0 e desvio padrão 1 é a Padronização.
